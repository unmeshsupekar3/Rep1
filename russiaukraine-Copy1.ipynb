```python
filepath3=r"C:\Users\unmes\Downloads\archive\sheet12.csv"
diff_df= pd.read_csv(filepath3)
diff_df.info()
```

```python
diff_df1=diff_df
diff_df1
```

```python
diff_df2=diff_df1.drop(columns=["Unnamed: 0","personnel","personnel*","POW","aircraft","helicopter","tank","APC","field artillery","MRL","drone","naval ship","anti-aircraft warfare","special equipment","vehicles and fuel tanks","cruise missiles"])
diff_df2.info()
```

```python
diff_df2.shape
```

```python
diff_df2=diff_df2.fillna(0)
diff_df2.isna().sum()
```

```python
diff_df2.info()
```

```python
diff_df2=diff_df2.astype({"D-special equipment":"int64","D-cruise missiles":"int64"})
diff_df2.info()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(diff_df2.corr(),annot=True) # To find correlation 
```

```python
diff_df2.corr()
```

```python
from scipy.stats import pearsonr
list1 = diff_df2['D-personnel']
list2 = diff_df2['D-APC']    #'D-aircraft','D-helicopter','D-tank',
 
# Apply the pearsonr()
corr, _ = pearsonr(list1, list2)
print('Pearsons correlation: %.3f' % corr)
 
```
